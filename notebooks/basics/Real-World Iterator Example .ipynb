```python
class FileLineReader:
    def __init__(self, filename):
        self.file = open(filename, "r")

    def __iter__(self):
        return self

    def __next__(self):
        line = self.file.readline()
        if line:
            return line.strip()
        self.file.close()
        raise StopIteration

for line in FileLineReader("data.csv"):
    print(line)
```
