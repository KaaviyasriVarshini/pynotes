```python
#8. Write a program that:
#Takes two numbers as input from the user.
#Computes and prints the sum, difference, product, quotient, remainder, and power of the numbers.

numbers = input("Enter two numbers: ").split()

a = int(numbers[0])
b = int(numbers[1])

print(a + b) #sum
print(a - b) #difference
print(a * b) #product
print(a / b) #quotient
print(a % b) #remainder
print(a ** b) #power of the numbers
```
